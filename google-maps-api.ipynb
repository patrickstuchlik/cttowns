Also can limit the number of times we use the API by going to the town master list first

```python
url = 'https://github.com/patrickstuchlik/cttowns/raw/main/towns.txt'
cttowns = pd.read_csv(url,header=None)
cttowns.rename(columns={0:'cttowns'},inplace=True)
cttowns.head()
```

```python
assert len(cttowns) == 169
```

```python
maps = pd.DataFrame({
    'address' : ['1245 chpael st new hvaen ct','83 Windham St, Willimantic, CT 06226','5 Sinawoy Rd, Cos Cob, CT 06807','fair haven, ct','10 N Water St, Norwalk, CT 06854'],
    'city':['new hvaen','Willimantic','Cos Cob','fair haven','Norwalk']
    }
)
maps.head()
```

```python
maps['cttown'] = ''

for ind in maps.index:
  if maps['city'][ind] in list(cttowns['cttowns']):
    maps['cttown'][ind] = maps['city'][ind]
  else:
    maps['cttown'][ind] = next(item for item in gmaps.geocode(maps['address'][ind])[0]['address_components'] if item['types']==['locality', 'political'])['long_name']
```
